```python
data_url = 'http://bit.ly/2cLzoxH'
# read data from url as pandas dataframe
data = pd.read_csv(data_url)
print(data.head(3))
```

```python
data.dropna(how="any",inplace = True)
data.info()
```

```python
# Standard imports 
from bokeh.io import output_notebook, show
output_notebook()
```

```python
# Specific import
from bokeh.io import output_file, show
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper
from bokeh.palettes import plasma
from bokeh.plotting import figure
from bokeh.transform import transform
```

```python
## solamente queremos un pais 
data_afg = data [data ['country']=='Afghanistan']
data_afg.shape
data_afg.head(12)
```

```python
list_y = data_afg['lifeExp'] # extectativa de vida 
list_x = data_afg['year'] # anos 
desc = [str(i) for i in list_y]

source = ColumnDataSource(data=dict(x=list_x, y=list_y, desc=desc))
hover = HoverTool(tooltips=[
    ("(x,y)", "(@x, @y)"),
    ('desc', '@desc'),
])
mapper = LinearColorMapper(palette=plasma(256), low=min(list_y), high=max(list_y))

p = figure(plot_width=400, plot_height=400, tools=[hover], title="Life Expectancy")
p.circle('x', 'y', size=10, source=source,
         fill_color=transform('y', mapper))

output_file('test.html')
show(p)
```
